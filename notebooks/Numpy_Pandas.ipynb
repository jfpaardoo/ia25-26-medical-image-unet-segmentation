## Numpy

El núcleo del paquete Numpy lo conforma el objeto `ndarray`, que es una implementación del tipo de dato _array_ multidimensional (es decir, una colección de elementos identificados por uno o más índices). Las instancias de `ndarray` se diferencian de las secuencias estándar de Python fundamentalmente en que:

- Tienen un tamaño fijo.
- Se requiere que todos sus elementos sean del mismo tipo.

Estas restricciones permiten a Numpy proporcionar implementaciones eficientes de muchas operaciones matemáticas, facilitando de esta manera trabajar con grandes conjuntos de datos.

Para una mayor legibilidad del código, a la hora de importar el paquete Numpy se suele adoptar el convenio de asociarle el nombre `np`.

```python
import numpy as np
```

### Creación de _arrays_ especificando los elementos

La función `array` permite crear _arrays_ a partir de secuencias estándar de Python anidadas.

```python
a = np.array([1.2, 3.5, 5.1, 4.8, 2.3])
a
```

```python
b = np.array([(4, -3, 0, -3), (3, -5, 1, 5), (-4, 6, -10, 9)])
b
```

**Nota**: es importante proporcionar los datos dentro de una secuencia y no como argumentos separados.

```python
np.array(1.2, 3.5, 5.1)
```

### Tamaño de un _array_

El tamaño de un _array_ viene determinado por su forma, que se guarda en el atributo `shape` y que es una tupla que recoge el tamaño de cada una de sus dimensiones. Cada una de estas dimensiones recibe en Numpy el nombre de eje (`axis`).

```python
print(f'Número de ejes del array a: {a.ndim}')
print(f'Forma del array a: {a.shape}')
print(f'Número de elementos del array a: {a.size}')
```

```python
print(f'Número de ejes del array b: {b.ndim}')
print(f'Forma del array b: {b.shape}')
print(f'Número de elementos del array b: {b.size}')
```

En Numpy, un vector hace referencia a un _array_ unidimensional, una matriz a un _array_ bidimensional y un tensor a un _array_ multidimensional.

El método `reshape` proporciona un _array_ con la nueva forma especificada. Hay que asegurarse de que la nueva forma es compatible con la cantidad de elementos del array. Si para un eje se especifica el valor `-1`, el tamaño de ese eje se calcula automáticamente.

```python
b.reshape(4, 3)
```

```python
b.reshape(2, 5)
```

```python
c = b.reshape(2, 2, -1)  # Le ponemos nombre para usarlo en ejemplos posteriores
c
```

### Creación de _arrays_ especificando el tamaño

Numpy ofrece diversas funciones para crear _arrays_ con una forma determinada, pero cuyos elementos actúan como marcadores de posición que podrán cambiarse posteriormente.

```python
np.zeros(3)
```

```python
np.identity(3)
```

```python
np.ones((3, 4))
```

```python
np.full((2, 5), 3.8)  # Rellena el array con el valor especificado
```

```python
np.empty((1, 2, 3))  # Los elementos del array son valores aleatorios que deberán modificarse convenientemente
```

### Tipo de dato de un _array_

El tipo de dato de los elementos de un _array_ está guardado en el atributo `dtype`.

```python
print(f'Tipo de dato de los elementos del array a: {a.dtype}')
print(f'Tipo de dato de los elementos del array b: {b.dtype}')
```

Numpy tiene implementados diferentes [tipos de datos estándar](https://numpy.org/doc/stable/reference/arrays.scalars.html#built-in-scalar-types) que determinan, entre otras propiedades, la cantidad de memoria que usan.

Se puede indicar el tipo de dato de sus elementos al construir un _array_.

```python
np.array([(1, 2, 3), (4, 5, 6)], dtype='float64')
```

```python
np.ones((3, 4), dtype='int8')
```

La mejor manera de cambiar el tipo de dato es crear una copia del _array_ mediante el método `astype`.

```python
a.astype('U')
```

### Selección de elementos de un _array_

La selección de elementos de un _array_ unidimensional se puede realizar de manera análoga a las listas de Python, usando índices y rebanadas (_slices_) de la forma `i:j:k`.

**Nota**: recuérdese que en Python las secuencias se enumeran empezando con el índice 0 y que los índices negativos toman como referencia el final de la secuencia.

```python
a[1]  # Selecciona el elemento de índice 1 (es decir, el segundo elemento de la secuencia)
```

```python
a[-1]  # Selecciona el último elemento
```

```python
a[0:2]  # El intervalo de índices definido por una rebanada es cerrado por la izquierda y abierto por la derecha
```

```python
a[:2]  # El índice inicial por defecto es el 0
```

```python
a[1:]  # El índice final por defecto es la longitud de la secuencia
```

```python
a[::2]  # Selecciona todos los elementos de índice par
```

Para los _arrays_ multidimensionales hay que especificar una expresión de selección por cada eje, separadas por comas. Para seleccionar todos los índices de un eje se usa la expresión `:` y para seleccionar todos los índices de varios ejes consecutivos se usa la expresión `...`.

```python
b[:2, -2:]  # Selecciona los elementos de las dos primeras filas y dos últimas columnas
```

```python
b[1, :]  # Selecciona la segunda fila como un array unidimensional
```

```python
b[[1], :]  # Para mantener la forma del array la fila debe seleccionarse mediante una secuencia
```

```python
c[0, :, :]  # Selecciona todos los elementos de índice 0 en el primer eje
```

```python
c[0, ...]  # Selecciona todos los elementos de índice 0 en el primer eje
```

```python
c[..., 0]  # Selecciona todos los elementos de índice 0 en el tercer eje
```

```python
c[0]  # La selección [i] es equivalente a la selección [i, ...]
```

También se pueden seleccionar elementos mediante secuencias de valores lógicos: se seleccionan los índices que se corresponden con el valor `True` y no se seleccionan los que se corresponden con el valor `False`. Lo habitual es usar esta manera de seleccionar proporcionando directamente una expresión que representa la condición que debe cumplirse para que un elemento sea seleccionado.

```python
a[a < 4]  # Selecciona los elementos de a menores que 4
```

```python
b[b % 2 == 0]  # Selecciona los elementos pares de b
```

```python
d = np.array(['No', 'Sí', 'Sí', 'No', 'Sí'])
a[d == 'Sí']  # Selecciona los elementos de a indicados por d
```

### Modificación de elementos de un _array_

Las expresiones de selección de elementos de un _array_ no dan como resultado un nuevo _array_, sino que proporcionan lo que se conoce como una vista del _array_ original. La combinación de estas vistas con el operador de asignación permiten modificar los elementos del _array_.

```python
a[0] = 0
a
```

```python
b[b % 2 == 0] += 1
b
```

Las operaciones de Numpy que actúan sobre un _array_ proporcionan como resultado una vista de este si es posible, permitiendo, en ese caso, modificar el _array_ original.

```python
b.reshape(4, 3)[0, 0] = 0
b
```

```python
b.astype('float64')[0, 0] = 5  # El método astype devuelve una copia del array original
b
```

El método `copy` nos permite obtener una copia, en lugar de una vista de un _array_.

```python
b.reshape(4, 3).copy()[0, 0] = 5
b
```

### Funciones universales

En Numpy, las [operaciones matemáticas](https://numpy.org/doc/stable/reference/routines.math.html) están implementadas mediante _funciones universales_. Una de las propiedades de este tipo de funciones es que están vectorizadas. Esto quiere decir que al aplicarlas a un _array_, entonces se aplican a cada elemento, devolviendo los resultados en un nuevo _array_.

```python
np.sqrt(a)  # Calcula la raíz cuadrada de cada elemento de a
```

```python
b + b  # Suma la matriz b consigo misma, elemento a elemento
```

Estas vectorizaciones se realizan, por supuesto, mediante bucles, pero al estar estos implementados en el lenguaje C se obtiene una mayor eficiencia que con un bucle estándar de Python.

```python
%%time

[2 * x for x in range(100000)];  # El punto y coma final evita que se muestre el resultado de la expresión
```

```python
%%time

2 * np.arange(100000);  # La función arange de Numpy es el equivalente de la función range de Python
```

Muchas operaciones escalares están implementadas como métodos del objeto _array_ y, además, pueden aplicarse a lo largo de uno o más ejes.

```python
b.sum()  # Suma de todos los elementos de b
```

```python
b.min(axis=0)  # Mínimo de cada columna de b
```

```python
b.max(axis=1)  # Máximo de cada fila de b
```

## Pandas

El núcleo del paquete Pandas lo conforman dos tipos de objetos:

- El objeto `Series` implementa un _array_ unidimensional en el que cada elemento tiene asignado un nombre.
- El objeto `DataFrame` implementa una tabla bidimensional en el que cada fila y cada columna tienen asignado un nombre.

Para una mayor legibilidad del código, a la hora de importar el paquete Pandas se suele adoptar el convenio de asociarle el nombre `pd`.

```python
import pandas as pd
```

### Creación de _Series_

Para crear una instancia de `Series` basta proporcionar la secuencia de elementos y usar el argumento `index` para proporcionar los nombres de esos elementos.

```python
notas_examen = pd.Series([5.9, 7.6, 0.5, 9.4], index=['Ana', 'Blanca', 'Carlos', 'David'])
notas_examen
```

Aunque es más habitual crear la instancia a partir de un diccionario.

```python
notas_examen = pd.Series({'Ana': 5.9, 'Blanca': 7.6, 'Carlos': 0.5, 'David': 9.4})
notas_examen
```

Si no se proporcionan los nombres, entonces los elementos se numeran consecutivamente a partir de 0.

```python
pd.Series([5.9, 7.6, 0.5, 9.4])
```

Las instancias de `Series` son, en esencia, _arrays_ de Numpy y, en consecuencia, todos sus elementos deben ser del mismo tipo.

```python
print(f'Número de ejes: {notas_examen.ndim}')
print(f'Forma: {notas_examen.shape}')
print(f'Tamaño: {notas_examen.size}')
print(f'Tipo: {notas_examen.dtype}')
```

### Selección y modificación de elementos de _Series_

La selección y modificación de elementos de una instancia de `Series` se realiza habitualmente por su nombre. Para seleccionar los elementos por su índice se recomienda el uso del método `iloc`.

```python
notas_examen.iloc[2]  # La tercera nota
```

```python
notas_examen['Carlos']  # La nota de Carlos
```

```python
notas_examen[:3]  # Las tres primeras notas
```

```python
notas_examen[notas_examen >= 5]  # Las notas mayores o iguales a 5
```

```python
notas_examen.iloc[[3, 0]]  # La última y la primera nota
```

```python
notas_examen[['David', 'Ana']]  # O, equivalentemente, las notas de David y Ana
```

```python
notas_examen['Carlos'] = 3  # Se revisa la nota de Carlos
notas_examen
```

Se pueden usar también rebanadas (_slices_) de nombre, pero hay que tener en cuenta que, en ese caso, el extremo derecho está incluido.

```python
notas_examen['Ana':'Carlos']  # Las notas desde Ana hasta Carlos, ambos inclusive
```

### Operaciones con _Series_

Como _arrays_ de Numpy que son, una función universal aplicada a una instancia de `Series` se aplica cada uno de sus elementos.

```python
2 * notas_examen
```

Al operar con varias instancias, estas se alinean por nombre.

```python
notas_otro_examen = pd.Series({'David': 8.9, 'Ana': 6.4, 'Carlos': 0.9, 'Blanca': 5.7})
notas_otro_examen
```

```python
notas_examen + notas_otro_examen
```

### Creación de _DataFrames_

Para crear una instancia de `DataFrame` basta proporcionar los elementos en una estructura bidimensional y usar los argumentos `index` y `columns` para proporcionar los nombres de las filas y las columnas, respectivamente.

```python
notas = pd.DataFrame([['FEX8601', False, 5.9, 6.4],
                      ['RQI5072', False, 7.6, 5.7],
                      ['ICA5467', True,  3.0, 0.9],
                      ['SRF5964', False, 9.4, 8.9]],
                     index=['Ana', 'Blanca', 'Carlos', 'David'],
                     columns=['UVUS', 'Repetidor?', 'Examen 1', 'Examen 2'])
notas
```

Aunque es más habitual crear la instancia a partir de un diccionario de las columnas.

```python
notas = pd.DataFrame({'UVUS': ['FEX8601', 'RQI5072', 'ICA5467', 'SRF5964'],
                      'Repetidor?': [False, False, True, False],
                      'Examen 1': [5.9, 7.6, 3.0, 9.4],
                      'Examen 2': [6.4, 5.7, 0.9, 8.9]},
                     index=['Ana', 'Blanca', 'Carlos', 'David'])
notas
```

Si no se proporcionan los nombres de las filas y/o de las columnas, entonces se numeran consecutivamente a partir de 0.

```python
notas = pd.DataFrame([['FEX8601', False, 5.9, 6.4],
                      ['RQI5072', False, 7.6, 5.7],
                      ['ICA5467', True,  3.0, 0.9],
                      ['SRF5964', False, 9.4, 8.9]])
notas
```

Aunque, si es posible, se usarán los nombres de los datos proporcionados.

```python
notas = pd.DataFrame({'UVUS': ['FEX8601', 'RQI5072', 'ICA5467', 'SRF5964'],
                      'Repetidor?': [False, False, True, False],
                      'Examen 1': notas_examen,
                      'Examen 2': notas_otro_examen})
notas
```

Una instancia de `DataFrame` es, en esencia, un diccionario (ordenado) de sus columnas, donde cada una de estas es una instancia de `Series`. En consecuencia, todos los elementos de una columna deben ser del mismo tipo, pero este puede diferir entre las distintas columnas.

```python
print(f'Número de ejes: {notas.ndim}')
print(f'Forma: {notas.shape}')
print(f'Tamaño: {notas.size}')
print(f'Tipo:\n{notas.dtypes}')
```

### Selección y modificación de elementos de _DataFrames_

Usando el operador `[]`, las cadenas seleccionan columnas por su nombre y las rebanadas (_slices_) filas por su posición.

```python
notas['UVUS']  # Al seleccionar un único nombre el resultado es una Series
```

```python
notas[['Examen 1', 'Examen 2', 'Repetidor?']]  # Al seleccionar una secuencia de nombres el resultado es un DataFrame
```

```python
notas[1:3]  # Al seleccionar filas el resultado siempre es un DataFrame
```

No obstante lo anterior, en general es preferible usar los métodos optimizados de selección implementados en Pandas: `iat` e `iloc` para seleccionar por posición; `at` y `loc` para seleccionar por nombre.

```python
notas.iat[1, 2]  # iat es el método más eficiente para obtener un único valor por su posición
```

```python
notas.iloc[1, 2]  # Es equivalente a la expresión anterior, pero menos eficiente
```

```python
notas.iloc[[0, 1, 2], [2, 3]]  # Se pueden seleccionar más de una fila y más de una columna por sus índices
```

```python
notas.iloc[:3, 2:4]  # Se pueden usar rebanadas de índices
```

```python
notas.iloc[1]  # Un único índice proporciona la fila correspondiente como Series
```

```python
notas.iloc[:2]  # Una secuencia de índices proporciona las filas correspondientes como DataFrame
```

```python
notas.at['Blanca', 'Examen 1']  # at es el método más eficiente para obtener un único valor por su nombre
```

```python
notas.loc['Blanca', 'Examen 1']  # Es equivalente a la expresión anterior, pero menos eficiente
```

```python
notas.loc[['Ana', 'Blanca', 'Carlos'], ['Examen 1', 'Examen 2']]  # Se pueden seleccionar más de una fila y más de una columna por sus nombres
```

```python
notas.loc['Ana':'Carlos', 'Examen 1':'Examen 2']  # Se pueden usar rebanadas de nombres
```

```python
notas.loc['Blanca']  # Un único nombre proporciona la fila correspondiente como Series
```

```python
notas.loc['Ana':'Blanca']  # Una secuencia de nombres proporciona las filas correspondientes como DataFrame
```

Combinando selección con el operador de asignación se pueden modificar los elementos de un DataFrame.

```python
notas.loc['Carlos', 'Examen 2'] = 2.8
notas
```

Es incluso posible extender un DataFrame con nuevas filas y columnas.

```python
notas.loc['Elena'] = ['SLE3920', 'True', 6.0, 4.8]
notas
```

```python
notas.loc[:, 'Examen 3'] = [6.2, 8.0, None, 9.5, 6.9]  # El valor None usado para indicar un examen no realizado se guarda como un objeto NaN (Not a Number),
                                                       # que representa un número no conocido
notas
```

### Estadística descriptiva

Pandas proporciona una gran cantidad de [métodos para calcular estadísticos descriptivos](https://pandas.pydata.org/docs/reference/frame.html#computations-descriptive-stats) sobre una _Series_ o _DataFrame_. La mayoría de estos métodos, como por ejemplo `sum` (suma) y `mean` (media), calculan valores agregados que devuelven en un **nuevo** objeto de menor dimensión que el de partida. Sin embargo, algunos de ellos, como por ejemplo `cumsum` (sumas acumuladas) y `cumprod` (productos acumulados), proporcionan los resultados en un **nuevo** objeto de la misma dimensión que el de partida.

```python
notas.loc[:, 'Examen 1':'Examen 3'].sum()  # Suma de las notas de cada examen
```

```python
notas.loc[:, 'Examen 1':'Examen 3'].cumsum()  # Sumas acumuladas de las notas de cada examen
```

En general, estos métodos disponen de un argumento `axis` para indicar a lo largo de qué eje realizar la operación que corresponda. Si el valor de `axis` es 0 (el valor por defecto), la operación se realiza a lo largo de las filas, mientras que si el valor de `axis` es 1, la operación se realiza a lo largo de las columnas.

```python
notas.loc[:, 'Examen 1':'Examen 3'].mean(axis=1)  # Media de las notas de cada alumno
```

Todos los métodos disponen de un argumento `skipna` para indicar qué tratamiento se les da a los valores no conocidos. Si el valor de `skipna` es `True` (valor por defecto), entonces no se incluyen en los cálculos, mientras que sí se incluyen si su valor es `False`.

```python
notas.loc[:, 'Examen 1':'Examen 3'].mean(axis=1, skipna=False)  # Media de las notas de cada alumno (incluyendo los exámenes sin nota)
```

Una opción mejor es realizar lo que se conoce como una imputación de valores. Por ejemplo, para el cálculo de la media de las notas de cada alumno tiene más sentido considerar los valores desconocidos como 0, en lugar de ignorarlos.

```python
notas.loc[:, 'Examen 1':'Examen 3'].fillna(0).mean(axis=1)  # El método fillna sustituye los valores desconocidos por el valor especificado
```

### Lectura y escritura de ficheros de datos

Pandas implementa una serie de [funciones y métodos para la lectura y escritura de ficheros de datos](https://pandas.pydata.org/docs/user_guide/io.html) en distintos formatos.

Un caso habitual es el de disponer de un conjunto de datos guardado en un fichero con [formato CSV](https://es.wikipedia.org/wiki/Valores_separados_por_comas) (_comma separated values_). Para leer los datos del fichero en un _DataFrame_ se proporciona la función `read_csv`, mientras que para escribir los datos de un _DataFrame_ en el fichero se proporciona el método `to_csv`. Ambas funciones disponen de argumentos que permiten adaptar la lectura y escritura al formato concreto del fichero.

```python
notas = pd.read_csv('notas.csv',         # Ruta del fichero a leer
                    skiprows=3,          # Descartar las tres primeras líneas del fichero
                    index_col='Alumno',  # La columna Alumno contiene el nombre de las filas
                    na_values=-1)        # Tratar el -1 como valor no conocido
```

```python
notas.head()  # Primeras filas del DataFrame
```

```python
notas.tail()  # Últimas filas del DataFrame
```

```python
notas.loc[:, 'Nota final'] = notas.loc[:, 'Examen 1':'Examen 3'].fillna(0).mean(axis=1)  # Calculamos la nota final de cada alumno como la media de las notas de los exámenes
notas
```

```python
notas = notas.round({'Nota final': 1})  # Redondeamos la nota final a un decimal
notas
```

```python
notas.to_csv('notas_finales.csv',  # Ruta del fichero a escribir
             na_rep = '-1')        # Escribir los valores no conocidos como -1
```
